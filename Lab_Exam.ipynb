```python
import pandas as pd
```

```python
df=pd.read_csv('reviews_dataset.csv')
```

```python
df.head()
```

```python


import re

def remove_non_alphabetic_words(text):
  # Use regular expressions to remove non-alphabetic words
  words = re.findall(r'\b[a-zA-Z]+\b', text)
  return " ".join(words)

# Apply the function to the 'news' column
df['news'] = df['news'].apply(remove_non_alphabetic_words)
```

```python
df['news']
```

```python
# prompt: convert df['news'] in into lower

df['news'] = df['news'].str.lower()
df['news']
```

```python
print(df['type'].value_counts())
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python

#df['sentences'] = df['review'].apply(sent_tokenize)
df['news'] = df['news'].apply(word_tokenize)
```

```python
df.head()
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmitizer = WordNetLemmatizer()
def lemmitize_words(text):
    return " ".join([lemmitizer.lemmatize(word,pos='v') for word in text.split()])
```

```python

def lemmatize_words(tokens):
    return [lemmitizer.lemmatize(word,pos='v') for word in tokens]
    #return " ".join([lemmatizer.lemmatize(word) for word in tokens])

# Lemmatizing the tokenized words in the 'review' column
df['lemmatized_news'] = df['news'].apply(lemmatize_words)
```

```python
df.head()
```

```python

from sklearn.preprocessing import LabelEncoder

X = df.drop('news', axis=1)
X = df['lemmatized_news']
Y = df['type']

print(X)
print(Y)

encoder = LabelEncoder()
Y = encoder.fit_transform(Y)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()


X_train_g = [" ".join(words) for words in X_train]
X_test_g = [" ".join(words) for words in X_test]

X_train_vec = vectorizer.fit_transform(X_train_g)
X_test_vec = vectorizer.transform(X_test_g)
```

```python
rf = RandomForestClassifier(n_estimators=150)

rf.fit(X_train_vec,y_train)
y_pred = rf.predict(X_test_vec)

print (accuracy_score(y_test,y_pred))
print (confusion_matrix(y_test,y_pred))
```

```python
vectorizer = CountVectorizer(ngram_range=(2, 2)) # Use ngram_range to specify bigrams

X_trainbi = [" ".join(words) for words in X_train]
X_testbi = [" ".join(words) for words in X_test]

X_train_vec_bi = vectorizer.fit_transform(X_trainbi)
X_test_vec_bi = vectorizer.transform(X_testbi)
```

```python
rf = RandomForestClassifier(n_estimators=150)
rf.fit(X_train_vec_bi,y_train)
y_pred = rf.predict(X_test_vec_bi)

print (accuracy_score(y_test,y_pred))
print (confusion_matrix(y_test,y_pred))
```

```python
# prompt: apply trigram  on this

vectorizer = CountVectorizer(ngram_range=(3, 3)) # Use ngram_range to specify trigrams

X_traintri = [" ".join(words) for words in X_train]
X_testtri = [" ".join(words) for words in X_test]

X_train_vec_tri = vectorizer.fit_transform(X_traintri)
X_test_vec_tri = vectorizer.transform(X_testtri)

```

```python
rf = RandomForestClassifier(n_estimators=150)
rf.fit(X_train_vec_tri,y_train)
y_pred = rf.predict(X_test_vec_tri)

print (accuracy_score(y_test,y_pred))
print (confusion_matrix(y_test,y_pred))
```
